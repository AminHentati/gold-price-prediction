# 1.1 Importing Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from statsmodels.tsa.seasonal import seasonal_decompose

from dateutil import parser
from sklearn.linear_model import LinearRegression

from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error

import matplotlib.dates as mdates

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("goldstock.csv")
y = df['Close'] 

# We will start out by selecting features gold ETF features
gold_features = ['Open','High', 'Low', 'Volume']
X = df[gold_features]
X.head()
```

```python
# Define Model
gold_model = LinearRegression()

#Fit Model
gold_model.fit(X, y)

print("Making predicitons for the first 5 entries\n")
print(X.head())
print("\nThe predictions are:\n")
print(gold_model.predict(X.head()))
print("\nThe actual values are:\n")
print(y.head())
```

```python
from sklearn.metrics import mean_absolute_error

predicted_adj_close = gold_model.predict(X.head())
print(mean_absolute_error(y.head(),predicted_adj_close))

predicted_adj_close = gold_model.predict(X)
print(mean_absolute_error(y, predicted_adj_close))
```

```python
from sklearn.model_selection import train_test_split

# Partition data into training and validation groups
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
# Define a new model for training set
gold_model = LinearRegression()
# Fit model
gold_model.fit(train_X, train_y)

#get predicted prices on validation data
val_predictions = gold_model.predict(val_X)
print(mean_absolute_error(val_y,val_predictions))
```

```python
from sklearn.pipeline import Pipeline
gold_model = LinearRegression()

# Bundle preporcessing and modeling code in a pipeline
my_pipeline = Pipeline(steps=[('gold_model', gold_model)])
# Preprocessing of training data, fit model
my_pipeline.fit(train_X, train_y)

# Preprocessing of validation data, get predictions
preds = my_pipeline.predict(val_X)

# Evaluate the model
mae_score = mean_absolute_error(val_y, preds)
print('MAE:', mae_score)

# Display Model
sns.regplot(x=val_y, y=preds, line_kws={"color":"black"})
```

```python
import joblib

joblib.dump(my_pipeline, 'gold_price_model.pkl')
```

# 1.7 Seasonal Decomposition:


```python
# Ensure the data is sorted by date in ascending order
df = df.sort_index(ascending=True)

# Perform seasonal decomposition again
decomposition = seasonal_decompose(df['Close'], model='multiplicative', period=365)
decomposition.plot()
plt.show()
```

## Insight from Plot:
### Upward Trend:

The long-term trend of gold prices is upward, with a significant increase over time, especially towards the end of the period (2024). There's a slight decline in the earlier years (2015), followed by a steady increase towards 2024.
### Seasonal Pattern:

There is a clear seasonal pattern in gold prices, with regular fluctuations occurring annually, repeating every year.
### Effective Decomposition:

The residuals are well-behaved, suggesting that the decomposition model has effectively captured the trend and seasonal components, with no discernible patterns or trends remaining in the residuals.

# 2.1 ARIMA Model for Forecasting:

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error

# Split the data into train and test sets
train = df['Close'][:int(0.8 * len(df))]
test = df['Close'][int(0.8 * len(df)):]
```

# 2.2 Fit ARIMA model

```python
model = ARIMA(train, order=(5, 1, 0))
model_fit = model.fit()
```

# 2.3 Forecast

```python
forecast = model_fit.forecast(steps=len(test))
test.index = forecast.index  # Ensure test and forecast indices match
```

# 2.4 Plot the results

```python
plt.figure(figsize=(14, 7))
plt.plot(train, label='Train')
plt.plot(test, label='Test')
plt.plot(forecast, label='Forecast')
plt.legend()
plt.show()
```

# 2.5 Evaluate the model


```python
rmse = np.sqrt(mean_squared_error(test, forecast))
print(f'RMSE: {rmse}')
```

```python
import itertools

# Define the range of p, d, q values to try
p = d = q = range(0, 4)  # You can increase the range
pdq = list(itertools.product(p, d, q))

best_rmse = float('inf')
best_order = None

for order in pdq:
    try:
        model = ARIMA(train, order=order)
        model_fit = model.fit()
        forecast = model_fit.forecast(steps=len(test))
        rmse = np.sqrt(mean_squared_error(test, forecast))
        if rmse < best_rmse:
            best_rmse = rmse
            best_order = order
    except:
        continue

print(f"Best ARIMA order: {best_order} with RMSE: {best_rmse}")
```

```python


# Plotting the actual data (train, test) and forecasted values
plt.figure(figsize=(14, 7))
plt.plot(train, label='Train', color='blue')  # Training data
plt.plot(test, label='Test', color='green')  # Test data
plt.plot(forecast, label='Forecast', color='red')  # Forecasted data
plt.legend()
plt.title(f"Best ARIMA Model: {best_order} | RMSE: {best_rmse}")
plt.show()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Define the SARIMA model parameters (example)
p, d, q = 1, 1, 1  # Non-seasonal ARIMA order
P, D, Q, m = 1, 1, 1, 12  # Seasonal ARIMA order (12 for monthly data)

# Fit the SARIMA model
sarima_model = SARIMAX(train, order=(p, d, q), seasonal_order=(P, D, Q, m))
sarima_model_fit = sarima_model.fit()

# Forecast
forecast = sarima_model_fit.forecast(steps=len(test))

# Calculate RMSE
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(test, forecast))
print(f"SARIMA RMSE: {rmse}")
```

```python
import matplotlib.pyplot as plt

# Plotting the actual data (train, test) and forecasted values
plt.figure(figsize=(14, 7))
plt.plot(train, label='Train', color='blue')  # Training data
plt.plot(test, label='Test', color='green')  # Test data
plt.plot(forecast, label='Forecast', color='red')  # Forecasted data
plt.legend()
plt.title('SARIMA Model: Train, Test, and Forecast')
plt.show()
```
